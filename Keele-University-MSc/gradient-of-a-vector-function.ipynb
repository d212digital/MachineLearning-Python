```python
import numpy as np
import matplotlib.pyplot as plt

# Visualize the gradient of a vector function as a surface plane
 
# Define the range for x and y
x = np.linspace(-10,10,1000)
xv, yv = np.meshgrid(x, x, indexing='ij')
 
# Compute f(x,y) = x^2 + y^3
zv = xv**2 + yv**3
 
# Plot the surface
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(xv, yv, zv, cmap="viridis")
plt.show()
```

```python
# Visualize the gradient of a vector function as a vector field - larger values are 
# shown in yellow whilst lover values are shown in purple

# Define the range for x and y
x = np.linspace(-10,10,20)
xv, yv = np.meshgrid(x, x, indexing='ij')
 
# Compute the gradient of f(x,y)
fx = 2*xv
fy = 2*yv
 
# Convert the vector (fx,fy) into size and direction
size = np.sqrt(fx**2 + fy**2)
dir_x = fx/size
dir_y = fy/size
 
# Plot the surface
plt.figure(figsize=(6,6))
plt.quiver(xv, yv, dir_x, dir_y, size, cmap="viridis")
plt.show()
```
